```python
import numpy as np
import Mors as mors
import matplotlib.pyplot as plt
import os
```

```python
star = mors.Load('p05_stars/p05_m0.1.pickle')
star.PrintAvailableTracks()
```

```python
eff = 0.3 #efficiency coefficient. The fraction of energy from the star that is used to evaporate atmosphere

masses = np.linspace(0.1,1.2,50)
FxHZ_50 = np.empty(len(masses))
FxHZ_5 = np.empty(len(masses))
FxHZ_95 = np.empty(len(masses))


def M_dot(eff,R_pl,M_pl):
    '''
    eff --> efficiency coefficient
    R_pl --> radius of the planet
    M_pl --> mass of the planet
    '''

    t = np.linspace(10,5000,10000)
    p05_FxHZ = np.empty(len(t))
    p50_FxHZ = np.empty(len(t))
    p95_FxHZ = np.empty(len(t))
    

    #load in the same mass star from each set of rotator speed groups
    p05 = mors.Load('p05_stars/p05_m0.5040816326530612.pickle')
    p50 = mors.Load('p50_stars/p50_m0.5040816326530612.pickle')
    p95 = mors.Load('p95_stars/p95_m0.5040816326530612.pickle')

    for i,age in enumerate(t):
        p05_FxHZ[i] = p05.Value(Age = age, Quantity='FxHZ')
        p50_FxHZ[i] = p50.Value(Age = age, Quantity = 'FxHZ')
        p95_FxHZ[i] = p95.Value(Age = age, Quantity = 'FxHZ')
        
        
    
    F_xuv = np.array([p05_FxHZ,p50_FxHZ,p95_FxHZ])
    
    
    R_p = R_pl * 6.37E6
    R_xuv = 1.1*R_p
    G = 6.67E-11
    M_p = M_pl * 5.97E24
    M_a = M_p * 10**-3
    Fx = F_xuv*10000

    
    m_dot = eff*Fx*np.pi*R_xuv**2*R_p/G/M_p

    #percentage of mass remaining
    M = (m_dot*t + M_a)
    
    
    plt.plot(t,M[0])
    plt.plot(t,M[1])
    plt.plot(t,M[2])
    
    plt.xscale('log')
    plt.yscale('log')
    
```

```python
M_dot(0.15,1,0.3)
```

```python
def massLossRate(efficiency, M_pl, R_pl):
    #efficiency between 0-1
    #M_pl in fraction of earth
    #R_pl in fraction of earth

    #need to get M_pl in actual mass
    Mp = M_pl * 5.97E24

    #same for R_pl
    Rp = R_pl * 6.37E6

    #need flux over course of time for each case of rotator
    p05 = mors.Load('p05_stars/p05_m0.5040816326530612.pickle')
    p50 = mors.Load('p50_stars/p50_m0.5040816326530612.pickle')
    p95 = mors.Load('p95_stars/p95_m0.5040816326530612.pickle')

    #need age range
    t = np.linspace(10,5000,10000)

    p05_FxHZ = np.empty(len(t))
    p50_FxHZ = np.empty(len(t))
    p95_FxHZ = np.empty(len(t))
    
    for i,age in enumerate(t):
        p05_FxHZ[i] = p05.Value(Age = age, Quantity='FxHZ')
        p50_FxHZ[i] = p50.Value(Age = age, Quantity = 'FxHZ')
        p95_FxHZ[i] = p95.Value(Age = age, Quantity = 'FxHZ')

    #need radius of atmosphere
    R_atm = Rp*1.1

    #gravitational constant
    G = 6.67E-11
    
    
    Mdot05 = np.pi*efficiency*p05_FxHZ/1000*R_atm**2*Rp/(G*Mp)*3.171E8 #divide by 1000 to get flux in terms of kg instead of g, multiply by 3.171E8 to get in terms of kg/year
    #Mdot50 = np.pi*efficiency*p50_FxHZ*R_atm**2*Rp/(G*Mp)
    #Mdot95 = np.pi*efficiency*p95_FxHZ*R_atm**2*Rp/(G*Mp)
    
    #plt.plot(t,Mdot05)
    #plt.plot(t,Mdot50)
    #plt.plot(t,Mdot95)
    #plt.title('Mass loss rate vs time')

    #make t0 = 10 so t=0 at 10
    t0 = np.linspace(0,4990,10000)*10000000
    M05 = ((Mp-Mdot05*t0)/Mp)*100
    #print(M05)
    plt.plot(t,M05)
    plt.yscale('log')
    plt.xscale('log')
    plt.title('%mass remaining over time')
    plt.xlabel('Age (10Myr-5Gyr)')
    plt.ylabel('% mass')
    
```

```python
massLossRate(0.15,0.3,0.5)
```
