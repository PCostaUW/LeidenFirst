```python
import nbimporter
import Functions as fun
import numpy as np
import matplotlib.pyplot as plt
import Mors as mors
import pandas as pd
```

```python
ages = np.linspace(50,5000,10000)
data = pd.read_csv('updated.csv')
masses,chz = data['Mass'],data['CHZ']
```

```python
x1 = []
x2 = []
for mass in masses:
    x1tmp,x2tmp = fun.findHZ(mass,ages)
    x1.append(x1tmp)
    x2.append(x2tmp)
```

```python
chz = (np.array(x1)+np.array(x2))/2

plt.plot(x1,masses,'.',label='Recent Venus')
plt.plot(x2,masses,'.',label='Early Mars')
plt.plot(chz,masses,linestyle='-',label='CHZ')
plt.ylim((0.3,1.0))
plt.xlabel('Distance from star (AU)')
plt.ylabel(f'Mass of star ($M_\odot$)')
plt.title('Bounds of the continuously habitable zone as a function of stellar mass')
plt.legend()
```
