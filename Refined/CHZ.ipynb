```python
import nbimporter
import Functions as fun
import numpy as np
import matplotlib.pyplot as plt
import Mors as mors
import pandas as pd
```

```python
ages = np.linspace(50,5000,10000)
data = pd.read_csv('updated.csv')
masses,chz = data['Mass'],data['CHZ']
```

```python
x1 = []
x2 = []
for mass in masses:
    x1tmp,x2tmp = fun.findHZ(mass,ages)
    x1.append(x1tmp)
    x2.append(x2tmp)
```

```python
chz = (np.array(x1)+np.array(x2))/2

plt.figure(facecolor='gainsboro',edgecolor='blue')
plt.plot(x1,masses,'.',label='Recent Venus',color='goldenrod',markersize=8)
plt.plot(x2,masses,'.',label='Early Mars',color='firebrick',markersize=8)
plt.plot(chz,masses,linestyle='-',label='CHZ values used',color='palegoldenrod',linewidth=5)
plt.fill_betweenx(masses,x1,x2,color='lightslategrey',label='CHZ')
plt.ylim((0.3,1.0))
plt.xlabel('Orbital Distance (AU)',fontsize=14)
plt.ylabel(f'Stellar Mass ($M_\odot$)',fontsize=14)
plt.title('Continuously Habitable Zone',fontsize=16)
plt.tick_params(top=True)
plt.legend()
```
